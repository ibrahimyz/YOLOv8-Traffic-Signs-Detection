```python
import os
import shutil
from collections import Counter
 
Path = "path_to_main_directory"  # change this to your main directory
# ── 1) Parameters ─────────────────────────────────────────
labels_dir = os.path.join(Path, "labels")      # the folder where .txt files live
 
images_dir = os.path.join(Path, "images")      # the folder where .jpg files live
 
"""control_path = Path + "/control20"
os.makedirs(control_path, exist_ok=True)"""
"""park_path = Nova_Drive_Path + "/4"
durpark_path = Nova_Drive_Path + "/7"
 
os.makedirs(park_path, exist_ok=True)  # Create the directory if it doesn't exist
os.makedirs(durpark_path, exist_ok=True)  # Create the directory if it doesn't exist"""
 
#os.makedirs(doner_kavsak_path, exist_ok=True)  # Create the directory if it doesn't exist
# If you want to count by name, we’ll use this Counter.
label_counter = Counter()
 
# If you also want to keep track of “unknown” IDs:
unknown_counter = Counter()
 
# ── 2) Iterate over every .txt in the labels folder ──────────────────────────
for fname in os.listdir(labels_dir):
    # Skip anything that does not end with .txt
    if not fname.endswith(".txt"):
        continue
 
    full_path = os.path.join(labels_dir, fname)
    # Sanity check: make sure it really is a file
    if not os.path.isfile(full_path):
        continue
 
    # ── 3) Open and read each line (YOLO format: "<class_id> <x> <y> <w> <h>") ───
    with open(full_path, "r") as f:
        for raw_line in f:
            line = raw_line.strip()
 
            # Skip empty lines
            if not line:
                continue
 
            parts = line.split()
            try:
                label_id = int(parts[0])
                if label_id in [25]:
                    print(f"Processing label ID {label_id} in file {fname}")

 
 
            except ValueError:
                # If it isn’t even an integer, just print a warning and skip
                print(f"[WARNING] Could not parse label ID in file {fname}: '{line}'")
                continue
 
```
